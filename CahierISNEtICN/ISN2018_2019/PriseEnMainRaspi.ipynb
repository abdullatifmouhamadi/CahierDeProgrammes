# Prise en main de la raspberry Pi3 ou Pi3+

## Première connexion :

Pour parvenir à utiliser la raspberry PI3 sur votre réseau il vous faut la connecter au réseau filaire de votre box. 

Une fois cela réalisé vous pouvez utiliser un petit logiciel de scan du réseau pour découvrir l'IP que votre BOX lui a donné. 

Vous pouvez aussi utiliser un câble hdmi pour la connecter sur un téléviseur et un clavier USB et une souris USB pour interagir avec le nanoordinateur. 

Vous pouvez utiliser la session Moon dont le mot de passe est "guest" sans les guillemets.

Vous pouvez enfin utiliser un client vnc comme [TightVNC Java Viewer](https://www.tightvnc.com/download/2.8.3/tvnjviewer-2.8.3-bin-gnugpl.zip). Une fois installé demandez une nouvelle connexion, le port à saisir est 5901 la session tsuser le mot de passe "raspife3" sans les guillemets.


## Utilisation de Jupyterhub :

Ce service est disponible et lancé au démarrage de la raspberry Pi. Vus devez utiliser firefox ou Chrome pour vous connecter à la raspi. 

l'adresse est : https://IPDeVotreRaspi:8000/

Vous allez obtenir une fenêtre de connexion avec login et mot de passe. 
Connectez vous en moon avec le mot de passe guest. 

![bannière de login](images/loginJupyterhub.png)

Vous disposez du Cahier de Programmes mutualisé par le projet Tremplin. Vous pouvez le mettre à jour en ouvrant une fenêtre de terminal 

![Menu déroulant "new"](images/NewTerminal.png)

Dans la fenêtre vérifiez que vous êtes dans le dossier CahierDeProgrammes et une fois que vous y êtes saisissez la l'invite de commande : 

```
raspildfISN0??: git pull

```

Pour commmencer à travailler, créez un nouveau dossier par exemple "MiniProjet2" puis dans ce dossier créez un notebook python3. Vous pouvez commencer à rédiger votre projet en utilisant les champs Markdown pour expliquer ce que vous faites et les champs python pour écrire et exécuter vos programmes.

## Accès à vos livrets xml en ligne 

vous devez aller paramétrer vos accès sur la base eXist mise à votre disposition sur mon serveur ainsi qu'au WebDav de mon serveur. Ceci dit une fois ces réglages faits vous pouvez aussi accéder à votre espace documentaire dans l'ENT. Il faut que vous ayez créé un dossier "media" et dans ce dossier les sous répertoires : 

>images
>
>fichiers
>
>videos
>
>audio
>
>scripts

Vous pourrez alors déplacer ce qui se trouve sur mon serveur dans votre "accès WebDav" dans l'ENT. C'est une solution qui vous rend indépendant de ma machine pour les medias que vous placez dans votre livret xml. 

### paramétrage d'un accès eXist.

Ouvrez les préférences d'oXygen puis sélectionnez le menu "sources de données" 

![Préférences d'oXygen](images/PreferencesOxygen.png)

Commencez par cliquer sur "Créer une connexion XML eXist-db"

Remplissez les champs comme sur l'illustration ci-dessous. Ne modifiez pas l'utilisateur ni le mot de passe.

![Popup Création connecteur eXist](images/PopupExist.png)

Dans les "Connexions" apparait un connecteur ... vous devez renseigner votre login et votre mot de passe 

login : pnomcomplet

mot de passe : p1nomcomplet2018

### paramétrage de l'accès au WebDav.

Toujours dans connexions créez un connecteur avec le "+" puis renseignez le popup.

Complétez avec les informations de la figure ci-dessous. login et mot de passe identiques à ceux de la connexion eXist.

![Popup WebDav](images/PopupWebDav.png)

Validez les changements et rendez-vous dans la page d'édition d'oXygen, choisissez en haut et à droite l'icone "base de données" et le cas échéant demandez dans le menu "fenêtre" à réinitialiser la mise en page pour retrouver la présentation vue au Lycée. 

Vous pouvez charger votre fichier xml .... si vous rajoutez des medias, veillez à utiliser la derniere partie pour copier / coller les modèles d'insertion des différents éléments utiles. Glissez déposez les medias dans le WebDav dans le dossier correspondant à leurs types.
